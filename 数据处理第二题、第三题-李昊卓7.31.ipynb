```python
#第三题
import numpy as np
import pandas as pd
import talib as ta
import matplotlib.pyplot as plt


data =  pd.read_excel('课件\sz50.xlsx',sheetname=5, index_col='datetime')

Series=data.iloc[:,0]

rolling_mean = Series.rolling(window=10).mean()
rolling_mean.name = "10day rolling mean of 600036.XSHG"

plt.figure(figsize=(15, 7))
Series.plot()
rolling_mean.plot()
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()



```
